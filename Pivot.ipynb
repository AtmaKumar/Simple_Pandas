```python
import pandas as pd
data=pd.read_csv("framingham.csv")
```

```python
data1=data.dropna()
```

```python
In Pivot we have to pass the column for Value on which we can use aggfunc like Max,min or mean and we also have to pass  
```

```python
pd.pivot_table(data1,values="age",index="male",aggfunc=[min,max])
```

```python
import numpy as np
pd.pivot_table(data1,values="age",index="male",aggfunc=[np.mean])
```

```python
import numpy as np
pd.pivot_table(data1,values="cigsPerDay",index=["male","currentSmoker"],aggfunc=[min,max,np.mean])
```

```python
pd.pivot_table(data1,values="age",index=["male","currentSmoker"],aggfunc=[min,max])
```

```python
import numpy as np
pd.pivot_table(data1,values=["age","cigsPerDay"],index=["male","currentSmoker"],aggfunc=[min,max,np.mean])
```

```python
import numpy as np
pd.pivot_table(data1,values=["age","cigsPerDay","heartRate"],index=["male","currentSmoker"],aggfunc=[min,max,np.mean])
```

```python
import numpy as np
pd.pivot_table(data1,values=["age","cigsPerDay","heartRate","glucose"],index=["male","currentSmoker"],aggfunc=[min,max,np.mean])
```

```python
import numpy as np
pd.pivot_table(data1,values=["age","cigsPerDay","heartRate","glucose"],index=["male","currentSmoker"],aggfunc=[np.mean])
```

```python
import numpy as np
pd.pivot_table(data1,values=["age","cigsPerDay","heartRate","glucose"],index=["male","currentSmoker","prevalentStroke"],aggfunc=[np.mean])
```

```python
pd.pivot_table(data1,values=["age","cigsPerDay"],index=["male","currentSmoker"],columns=["education"],aggfunc=[np.mean])
```
